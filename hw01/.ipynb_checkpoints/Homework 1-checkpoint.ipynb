## Part1. DFT and FFT


### Theoretical introduction: Definition of DFT


Let $x\in \mathbb{C}^N$ (complex vector space) be a vector of length $N\in\mathbb{N}$. In the audio signal context, $x$ can be interpreted as a discrete signal with samples $x(0), x(1), ..., x(N-1)$. Note that we start indexing with the index $0$. The **discrete Fourier transform** (DFT) of $x$ is defined by:

$$ X(k) := \sum_{n=0}^{N-1} x(n) \exp(-2 \pi i k n / N) $$

for $k \in [0:N-1]$. The vector $X\in\mathbb{C}^N$ can be interpreted as frequency representation of the time-domain signal $x$.   To obtain a geometric interpretation of the DFT, we define the vector $\mathbf{u}_k\in\mathbb{C}^N$ by

$$\mathbf{u}_k(n) :=  \exp(2 \pi i k n / N) = \cos(2 \pi k n / N) + i \sin(2 \pi k n / N)$$

for each $k \in [0:N-1]$. This vector can be regarded as a sampled version of the exponential function of frequency $k/N$. Then, the DFT can be expressed as inner products

$$ X(k) := \sum_{n=0}^{N-1} x(n) \overline{\mathbf{u}_k} = \langle x | \mathbf{u}_k \rangle$$

of the signal $x$ and the sampled exponential functions $\mathbf{u}_k$. The absolute value $|X(k)|$ indicates the degree of similarity between the signal $x$ and $\mathbf{u}_k$.

In the case that $x\in \mathbb{R}^N$ is a real-valued vector (which is always the case for our audio signal scenario), we obtain:

$$ X(k) := \langle x |\mathrm{Re}(\mathbf{u}_k) \rangle - i\langle x | \mathrm{Im}(\mathbf{u}_k) \rangle $$

The following plot shows an example signal $x$ compared with functions $\overline{\mathbf{u}_k}$ for two different frequency parameters $k$. The real and imaginary part of $\overline{\mathbf{u}_k}$ are shown in <font color='red'> red</font> and <font color='blue'> blue</font>, respectively.





### DFT Matrix

Being a linear operator $\mathbb{C}^N \to \mathbb{C}^N$, the DFT can be expressed by some $N\times N$-matrix. This leads to the famous DFT matrix $\mathrm{DFT}_N \in \mathbb{C}^{N\times N}$ matrix, which is given by

$$\mathrm{DFT}_N(n, k) = \mathrm{exp}(-2 \pi i k n / N)$$

for $n\in[0:N-1]$ and $k\in[0:N-1]$. Let $\rho_N:=\exp(2 \pi i / N)$ be the primitive $N^\mathrm{th}$ root of unity. Then 

$$\sigma_N:= \overline{\rho_N} = \mathrm{exp}(-2 \pi i / N)$$

also defines a primitive $N^\mathrm{th}$ root of unity. From the properties of exponential functions, one obtains that

$$ \sigma_N^{kn} = \mathrm{exp}(-2 \pi i / N)^{kn} = \mathrm{exp}(-2 \pi i k n / N)$$

From this, one obtains:

$$
\mathrm{DFT}_N =
\begin{pmatrix}
    1 & 1 & 1 & \dots  & 1 \\
    1 & \sigma_N & \sigma_N^2 & \dots  & \sigma_N^{N-1} \\
    1 & \sigma_N^2 & \sigma_N^4 & \dots  & \sigma_N^{2(N-1)} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & \sigma_N^{N-1} & \sigma_N^{2(N-1)} & \dots  & \sigma_N^{(N-1)(N-1)} \\
\end{pmatrix}
$$

In the following visualization, the real and imaginary part of $\mathrm{DFT}_N$ are shown, where the values are encoded by suitable colors. Note that the $k^\mathrm{th}$ row of $\mathrm{DFT}_N$ corresponds to the vector $\mathbf{u}_k$ as defined above.

````python
import numpy as np

def generate_matrix_dft(N, K):
    
    """
    
    Generate a DFT (discete Fourier transfrom) matrix
    
    Args:
        N: Number of samples
        K: Number of frequency bins

    Returns:
        dft: The DFT matrix
        
    """
    
    dft = np.zeros((K, N), dtype=np.complex128)
    
    for n in range(N):
        for k in range(K):
            dft[k, n] = np.exp(-2j * np.pi * k * n / N)
            
    return dft

generate_matrix_dft( 3, 3 )
````

Complete the function ``` discrete_fourier_transform```  below to compute the DFT.

````python
def discrete_fourier_transform(x):
    
    """
    
        x <- input signal
    
    """

    x = x.astype(np.complex128) # complex representation od the input signal
    
    N = len(x) # sample size
    
    dft_mat = generate_matrix_dft(N, N) # DFT matrix (square matrix)
    
    # Complete this line ...
    # define the inner product of input signal with DFT
    
    discrete_fourier_transform = np.dot( dft_mat, x )
    
    return discrete_fourier_transform

    
N = 64 # sample size
n = np.arange( N ) # sample index [0, ..., N]
k = 10 # frequency

x = np.cos( 2 * np.pi * (k*n/N) )

discrete_fourier_transform( x )
````

````python
"""

    where am i?

"""

PATH = os.getcwd() + '/'
audio_path = PATH + 'genres_audio/'

print(f'PATH: {PATH}')
print(f'audio path: {audio_path}')
````

````python
"""

    input audio

"""

audio_ss = sorted(os.listdir( audio_path ))

print(f'sample space: {len(audio_ss)}')
print(f'sample names example: \n{ np.random.choice( audio_ss, 5 ) }')

len_f = []
sc_f = []
rms_f = []
zcr_f = []

for s in tqdm(audio_ss):
    
    # y <- audio time series
    # sr <- spectogram magnitude
    y, sr = librosa.load( audio_path + s )
    
    # lenght
    len_ = len(y)
    len_f.append( [s, len_] )
    
    # spectral centroid
    sc = librosa.feature.spectral_centroid(y=y)[0]
    sc_f.append( [s] + list(sc) )
    
    # mel spectogram
    rms = librosa.feature.rms(y=y)[0]
    rms_f.append( [s] + list(rms) )
    
    # zero crossing rate
    zcr = librosa.feature.zero_crossing_rate(y=y)[0]
    zcr_f.append( [s] + list(zcr) )
    
    
len_df = pd.DataFrame( len_f, columns=['sample', 'len'] )

sc_df = pd.DataFrame( sc_f ).rename(columns={0: 'filename'})
sc_df = pd.merge(sc_df, genre_dataset, on=['filename'])

rms_df = pd.DataFrame( rms_f ).rename(columns={0: 'filename'})
rms_df = pd.merge(rms_df, genre_dataset, on=['filename'])

zcr_df = pd.DataFrame( zcr_f ).rename(columns={0: 'filename'})
zcr_df = pd.merge(zcr_df, genre_dataset, on=['filename'])
````

````python
gl = sorted(genre_dataset['genre'].unique())

for g in gl:
    
    aux = sc_df[ sc_df['genre'] == g ]
    aux = aux[ [x for x in list(aux) if x not in ['filename', 'genre']] ].values
    
    means = np.mean( aux, axis=0 )
    stds = np.std( aux, axis=0 )
    
    # Plot error bands using Seaborn
    sns.lineplot(x=range(len(means)), y=means, label=g)
    plt.fill_between(range(len(means)), means - stds, means + stds, alpha=0.2)

plt.title('Spectral Centroid')
plt.show()
````

Please add your answer here: 

**- Spectral Centrum** 

    Comparing the spectral centroid of different music genres, we can observe that classical music tends to have lower centroid values, while metal tends to have higher centroid values. The spectral centroid captures the center of gravity of the frequency content, and from the plot it can be observed that metal has a higher concentration of energy in high frequencies, while classical music has a higher concentration of energy in low frequencies.
    
    It holds with the general knowledge that because of the instruments used in classical music work principally with lower frequencies.
    
**- Root Mean Square**
    
    The RMS (Root Mean Square) value is related to the amplitude of an audio signal and can be seen as a measure of its overall loudness. From the plot, we can observe that hip hop tends to have the highest RMS values, while classical music tends to have lower RMS values.
    
    This can be because the classical music has a more consistent amplitude profile, with fewer sudden changes in amplitude.
    
**- Zero Crossing Rate**

    This plot shows a distribution of genres that is similar to that of the spectral centroid. This could be because the zero crossing rate is also affected by the frequency of the audio signal. It can be interpreted as a measure of the rhythm structure of the audio signal, where lower values (in this case, blues) indicate a legato style, while higher values suggest a more percussive style.
